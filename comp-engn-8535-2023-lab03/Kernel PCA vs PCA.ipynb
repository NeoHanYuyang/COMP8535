# Kernel PCA vs PCA for Circular Data
In this lab, we are going to explore the performence of PCA and Kernel PCA for data with nonlinear structure. This lab is based on the tutorial [here](https://scikit-learn.org/stable/auto_examples/decomposition/plot_kernel_pca.html#sphx-glr-auto-examples-decomposition-plot-kernel-pca-py)

```python
import matplotlib.pyplot as plt
import numpy as np
```

### The code below is provided to generate the random input data

```python
from sklearn.datasets import make_circles
n = 1000 # number of data points
input_X, circle_labels = make_circles(n_samples=n, factor=0.3, noise=0.05, random_state=0) # routine to generate random data with structure of two circles
input_X = input_X.T
plt.scatter(input_X[0, :], input_X[1, :], c = circle_labels) # plot the data points (the circle labels are not part of the data, just help us visualise which points are which)
```

## Let's now implement PCA and Kernel PCA
Instead of implementing the routines from scratch, let's use the `sklearn.decomposition` routines `PCA` and `KernelPCA`.
* For PCA see the documentation [here](https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.PCA.html#sklearn.decomposition.PCA)
* For Kernel PCA see the documentation [here](https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.KernelPCA.html)

```python
from sklearn.decomposition import PCA, KernelPCA
```

### For PCA:
1. Create a PCA transformer object using $k = 2$ principle components via `pca = PCA(n_components = 2)`
2. Use the transformer object's method function `y_pca = pca.fit_transform(input_X.T)` to perform PCA on the input data.
3. Plot the PCA result.


```python
# Your code here to perform PCA 
pca = PCA(n_components=2)
y_pca = pca.fit_transform(input_X.T)

# Plotting Code
y_pca = y_pca.T
plt.figure()
plt.scatter(y_pca[0, :], y_pca[1, :], c=circle_labels)
plt.ylabel("Principal component #1")
plt.xlabel("Principal component #0")
plt.title("Projection of data\n using PCA")
```

### For Kernel PCA:
1. Create a Kernel PCA transformer object using $k = 2$ principle components via `kernel_pca = KernelPCA(n_components = 2, kernel="rbf", gamma=1)`
2. Use the transformer object's method function `y_kernel_pca = kernel_pca.fit_transform(input_X.T)` to perform kernel PCA on the input data. 
3. Plot the Kernel PCA result.
4. In Step 2 above, change the `gamma=1` option to `gamma=2` and compare the output of Kernel PCA. Explore other kernels and parameter options that enable the data to be linearly seperated. See kernel options [here](https://scikit-learn.org/stable/modules/generated/sklearn.decomposition.KernelPCA.html) and discussion of kernels [here](https://scikit-learn.org/stable/modules/metrics.html#metrics)

```python
# Your code here to perform kernel PCA
kernel_pca = KernelPCA(n_components=2, kernel='rbf', gamma=2)
y_kernel_pca = kernel_pca.fit_transform(input_X.T)

# Plotting Code
y_kernel_pca = y_kernel_pca.T
plt.figure()
plt.scatter(y_kernel_pca[0, :], y_kernel_pca[1, :], c=circle_labels)
plt.ylabel("Principal component #1")
plt.xlabel("Principal component #0")
plt.title("Projection of data\n using KernelPCA")
```

```python
# Your code here to perform kernel PCA
kernel_pca = KernelPCA(n_components=2, kernel='rbf', gamma=1)
y_kernel_pca = kernel_pca.fit_transform(input_X.T)

# Plotting Code
y_kernel_pca = y_kernel_pca.T
plt.figure()
plt.scatter(y_kernel_pca[0, :], y_kernel_pca[1, :], c=circle_labels)
plt.ylabel("Principal component #1")
plt.xlabel("Principal component #0")
plt.title("Projection of data\n using KernelPCA")
```
